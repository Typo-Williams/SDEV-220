** What was the average (mean) BasePay of all employees per year? (2011-2014) ? **

```python
result1 = sal[(sal['Year'] == 2011) & sal['TotalPayBenefits'] > 0]
result2 = sal[(sal['Year'] == 2012) & sal['TotalPayBenefits'] > 0]
result3 = sal[(sal['Year'] == 2013) & sal['TotalPayBenefits'] > 0]
result4 = sal[(sal['Year'] == 2014) & sal['TotalPayBenefits'] > 0]

print("2011", result1['TotalPayBenefits'].max())
print("2012", result2['TotalPayBenefits'].max())
print("2013", result3['TotalPayBenefits'].max())
print("2014", result4['TotalPayBenefits'].max())
```

** How many unique job titles are there? **

```python
sal['JobTitle'].nunique()
```

** What are the top 5 most common jobs? **

```python
sal['JobTitle'].value_counts().head(5)
```

** How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?) **

```python
sum(sal[sal['Year']==2013]['JobTitle'].value_counts() == 1)
```

** How many people have the word Chief in their job title? (This is pretty tricky) **

```python
def chiefs(title):
    if 'chief' in title.lower().split():
        return True
    else:
        return False
```

```python
sum(sal['JobTitle'].apply(lambda x: chiefs(x)))
```

** Bonus: Is there a correlation between length of the Job Title string and Salary? **

```python
sal['title_len'] = sal['JobTitle'].apply(len)
```

```python
sal[['TotalPayBenefits', 'title_len']].corr()
```
